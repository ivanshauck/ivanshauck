The purpose of this project is to segment credit card customers based on their spending habits from the previous 6 months. The idea will be to see how the most effective clustering model breaks them up, and analyze the differences between groups to find some actionable ways to treat the different customer segments.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

Now I'll run a variety of different clustering models with different settings to see which one works best. I'll be using the silhouette score as my metric since I'm mainly looking to analyze similar groups of customers, and not necessarily base my analysis on a single feature, as an ARI score would do.

```python
X = cc.drop(columns=['CUST_ID'])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets, metrics
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=22)
kmeans.fit(X_std)
clusters = kmeans.predict(X_std)
print(metrics.silhouette_score(X_std,clusters,metric='euclidean'))
```

```python
from sklearn.cluster import AgglomerativeClustering
agg1 = AgglomerativeClustering(linkage='average',affinity='manhattan',n_clusters=3)
agg2 = AgglomerativeClustering(linkage='ward',affinity='euclidean',n_clusters=3)
agg3 = AgglomerativeClustering(linkage='complete',affinity='manhattan',n_clusters=3)
c1 = agg1.fit_predict(X_std)
c2 = agg2.fit_predict(X_std)
c3 = agg3.fit_predict(X_std)
print(metrics.silhouette_score(X_std,c1,metric='euclidean'))
print(metrics.silhouette_score(X_std,c2,metric='euclidean'))
print(metrics.silhouette_score(X_std,c3,metric='euclidean'))
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=2, min_samples=5)
scan = dbscan.fit_predict(X_std)
print(metrics.silhouette_score(X_std,scan,metric='euclidean'))
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm_cluster = GaussianMixture(n_components=2, random_state=123)
gm = gmm_cluster.fit_predict(X_std)
print(metrics.silhouette_score(X_std,gm,metric='euclidean'))
```

The first and second clusters have very similar customers, with the main difference being that the first group seems to get cash advances very frequently, and the second group makes fewer purchases, and has an overall lower credit limit. The third group makes purchases much more often, seems to have a higher credit limit, and seems to pay off their balance more frequently and have a higher percent of their balance paid off, now I'll confirm this by comparing some means and conducting a few t tests.

```python
purchases1 = xc.loc[xc['cluster']==0].PURCHASES
purchases2 = xc.loc[xc['cluster']==1].PURCHASES
purchases3 = xc.loc[xc['cluster']==2].PURCHASES
maxpur1 = xc.loc[xc['cluster']==0].ONEOFF_PURCHASES
maxpur2 = xc.loc[xc['cluster']==1].ONEOFF_PURCHASES
maxpur3 = xc.loc[xc['cluster']==2].ONEOFF_PURCHASES
advance1 = xc.loc[xc['cluster']==0].CASH_ADVANCE
advance2 = xc.loc[xc['cluster']==1].CASH_ADVANCE
advance3 = xc.loc[xc['cluster']==2].CASH_ADVANCE
clim1 = xc.loc[xc['cluster']==0].CREDIT_LIMIT
clim2 = xc.loc[xc['cluster']==1].CREDIT_LIMIT
clim3 = xc.loc[xc['cluster']==2].CREDIT_LIMIT
```

Let's take a look at the distributions for these variables, not as a metric for comparing them, but simply to see what a/b test would be appropriate.

```python
plt.figure(figsize=(15,15))
plt.hist(purchases1,alpha=0.5)
plt.hist(purchases2,alpha=0.5)
plt.hist(purchases3,alpha=0.5)
plt.title('purchases distributions')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.hist(maxpur1,alpha=0.5)
plt.hist(maxpur2,alpha=0.5)
plt.hist(maxpur3,alpha=0.5)
plt.title('maximum purchase distributions')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.hist(advance1,alpha=0.5)
plt.hist(advance2,alpha=0.5)
plt.hist(advance3,alpha=0.5)
plt.title('total cash advance amount distributions')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.hist(clim1,alpha=0.5)
plt.hist(clim2,alpha=0.5)
plt.hist(clim3,alpha=0.5)
plt.title('credit limit distributions')
plt.show()
print(stats.describe(clim1))
print(stats.describe(clim2))
print(stats.describe(clim3))
```

Number of purchases, purchase amount, and cash advances all have distributions that do not look normal. For this reason I'll conduct a Kruskal Wallis test, although this test can be used for more than two groups, I'll conduct it between each group in pairs. This is to confirm that each group differs meaningfully from the the other two. This will help us understand if comparing the difference in means of the groups has value.

WIth regards to credit limit, aside from the second cluster's kurtosis being a tad above three, it looks as if all three distributions have skewness and kurtosis that fall within the limit of a normal distribution, so for these we can conduct a regular t test.

```python
print(stats.kruskal(purchases1,purchases2))
print(stats.kruskal(purchases1,purchases3))
print(stats.kruskal(purchases2,purchases3))
```

```python
print(stats.kruskal(maxpur1,maxpur2))
print(stats.kruskal(maxpur1,maxpur3))
print(stats.kruskal(maxpur2,maxpur3))
```

```python
print(stats.kruskal(advance1,advance2))
print(stats.kruskal(advance1,advance3))
print(stats.kruskal(advance2,advance3))
```

```python
print(stats.ttest_ind(clim1,clim2))
print(stats.ttest_ind(clim1,clim3))
print(stats.ttest_ind(clim2,clim3))
```

According to these tests every one of the groups differs meaningfully from the others, so we can safely compare means.
